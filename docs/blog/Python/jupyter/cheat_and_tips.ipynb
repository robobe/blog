## inline img
Using cv2 to read and manipulate image 
Show the image inline the notebook
Using `matplotlib` to show image

```python
import cv2
from  matplotlib import pyplot as plt

image = cv2.imread("../images/test.png")
# Convert from BGR to RGB
image = image[:,:,::-1] 
plt.imshow(image)
plt.show()
```
