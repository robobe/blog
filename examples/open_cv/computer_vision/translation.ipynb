```python
import cv2
import numpy as np
from  matplotlib import pyplot as plt
```

```python
image = cv2.imread("images/tomto.jpg")
image.shape
```

```python
height, width = image.shape[:2]
tx = 100
ty = 100
T = np.float32([[1, 0, tx], [0, 1, ty]])
tmp = np.array([[1,2,3],[4,5,6],[4,5,6]], dtype=np.float32)
print(tmp.shape)
print(tmp.size)
print(tmp.dtype)
```

```python
# the image using the matrix, T
img_translation = cv2.warpAffine(image, T, (width, height))
```

```python
image = img_translation[:,:,::-1] 
plt.imshow(image)
plt.show()
```
